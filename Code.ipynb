```python
import pandas as pd

df = pd.read_csv("YouTubeDataset_withChannelElapsed.csv")
df_unique_values = (
    df
    .groupby("videoId", as_index=False)
    .agg({
        "videoViewCount": "sum",
        "videoLikeCount": "sum",
        "videoDislikeCount": "sum",
        "VideoCommentCount": "sum"
    })
)
```

```python
#Question 1

average_likes = df_unique_values["videoLikeCount"].mean()
print(round(average_likes,4))

```

```python
#Question 2

df_engagement_rate = df_unique_values.copy()

df_engagement_rate["engagement_rate"] = (df_unique_values['videoLikeCount'] + df_unique_values['videoDislikeCount'] + df_unique_values['VideoCommentCount'])/df_unique_values['videoViewCount']

average_engagement_rate = df_engagement_rate["engagement_rate"].mean()

print(round(average_engagement_rate,4))
```

```python
#Question 3

byte_mb = 500/(1024**2)
kb_mb = 200/1024

result = byte_mb + kb_mb

print(round(result,4))
```

```python
#Question 4
df_engagement = df_unique_values.copy()
df_engagement['engagement'] = df_unique_values['videoLikeCount'] + df_unique_values['videoDislikeCount'] + df_unique_values['VideoCommentCount']


total_data_transfer = (df_engagement['engagement'].sum() * result)

print(round(total_data_transfer,4))
```

```python
#Question 5
view_per_min = (df_unique_values['videoViewCount'].sum()/(5*365*24*60))
print(f"{round(view_per_min,4)} number_of_videos/min ")
```

```python
#Question 6

data_per_min = view_per_min * result
print(f"{round(data_per_min,4)} MB/min")
```
